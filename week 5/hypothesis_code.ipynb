```python
# hypothesis code
import math
n = 100
x = 60
p_hat = x / n
SE = math.sqrt(p_hat * (1 - p_hat) / n)
z = 1.96
lower = p_hat - z * SE
upper = p_hat + z * SE
print("Sample proportion:", p_hat)
print("Confidence Interval:", (lower, upper))
p0 = 0.50
z_test = (p_hat - p0) / math.sqrt(p0 * (1 - p0) / n)
print("z value:", z_test)
if abs(z_test) > 1.96:
    print("Reject H0")
else:
    print("Do not Reject H0")
```
